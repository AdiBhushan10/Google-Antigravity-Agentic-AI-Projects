### Generative AI Application Using Agents and GROQ

```python
# initialise openai model using groq model provider/ groq is free but openai directly has cost
from langchain.chat_models import init_chat_model
model=init_chat_model("openai/gpt-oss-120b",model_provider="groq")
model
```

```python
response=model.invoke("Write me a quick 100 words intro on yourself")
print(response)
response.content
```
